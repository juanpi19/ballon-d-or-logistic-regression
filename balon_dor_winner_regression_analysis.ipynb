# Data Collection

### Team Stats

```python
# Scraping Team Stats from all 5 major european leagues from 2000-2001 to the most recent season

seasons = ['2000-2001', '2001-2002', '2002-2003', '2003-2004', '2004-2005', '2005-2006', '2006-2007', '2007-2008',
          '2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014', '2014-2015', '2015-2016',
          '2016-2017', '2017-2018', '2018-2019', '2020-2021', '2021-2022', '2022-2023']

team_data = pd.DataFrame()
columns = ['Rk', 'Squad', 'Country', 'LgRk', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD','Pts', 'Pts/MP', 
           'xG', 'xGA', 'xGD', 'xGD/90', 'Attendance','Top Team Scorer', 'Goalkeeper', 'season']

for i in columns:
    team_data[i] = np.nan

for season in seasons:
    df = pd.read_html(f'https://fbref.com/en/comps/Big5/{season}/{season}-Big-5-European-Leagues-Stats')[0]
    df['season'] = season
    team_data = pd.concat([team_data, df])
```

# Cleaning The Data

```python
# Team Data DataFrame Cleaning
team_data['Top Team Scorer'] = [i.split('-')[0].strip() for i in team_data['Top Team Scorer']]
team_data['Country'] = [i.split(' ')[1] for i in team_data['Country']]

# Player Data DataFrame Cleaning
player_data['Comp'] = [i.split()[0] for i in player_data['Comp']]
player_data.drop(['Matches1', 'Matches2'], axis=1, inplace=True) # dropping two columns: Matches1 and Matches 2

# Renaming Columns
team_data.columns = [i + "_team" for i in team_data.columns]
player_data.columns = [i + "_player" for i in player_data.columns]
wc_df.columns = [i + "_wc" for i in wc_df.columns]
ca_df.columns = [i + "_ca" for i in ca_df.columns]
ec_df.columns = [i + "_ec" for i in ec_df.columns]
cl_df.columns = [i + "_cl" for i in cl_df.columns]
ballon_df.columns = [i + "_ballon" for i in ballon_df.columns]
```

# Loading Data

```python
team_data = pd.read_csv('team_data.csv')
player_data = pd.read_csv('player_data.csv')
wc_df = pd.read_csv('wc_df.csv')
ca_df = pd.read_csv('ca_df.csv')
ec_df = pd.read_csv('ec_df.csv')
cl_df = pd.read_csv('cl_df.csv')
ballon_df = pd.read_csv('ballon_df.csv')


team_data.drop('Unnamed: 0', axis=1, inplace=True)
player_data.drop('Unnamed: 0', axis=1, inplace=True)
wc_df.drop('Unnamed: 0', axis=1, inplace=True)
ca_df.drop('Unnamed: 0', axis=1, inplace=True)
ec_df.drop('Unnamed: 0', axis=1, inplace=True)
cl_df.drop('Unnamed: 0', axis=1, inplace=True)
ballon_df.drop('Unnamed: 0', axis=1, inplace=True)
```

### Removing Unnecessary Columns

```python
player_data.drop(['min90_Gls_player', 
                  'min90_Ast_player', 
                  'min90_G+A_player',
                  'min90_G-PK_player', 
                  'min90_G+A-PK_player', 
                  'min90_xG_player',
                  'min90_xAG_player', 
                  'min90_xG+xAG_player', 
                  'min90_npxG_player',
                  'min90_npxG+xAG_player',
                  'xG_player',
                  'npxG_player',
                  'xAG_player',
                  'npxG+xAG_player',
                  'PrgC_player',
                  'PrgP_player',
                  'PrgR_player'], axis=1, inplace=True)

team_data.drop(['xG_team', 
                'xGA_team', 
                'xGD_team', 
                'xGD/90_team'], axis=1, inplace=True)
```

# Joining Datasets

```python
# Team data
df = pd.merge(player_data, 
         team_data, 
         how='outer', 
         left_on=['Squad_player', 'season_player'],
         right_on=['Squad_team', 'season_team'])

# World Cup
df = pd.merge(df, 
              wc_df, 
              how='left', 
              left_on='season_team', 
              right_on='season_wc')

# wc_df cleaning
df['is_wc_yr'] = 'no'
df.loc[df['Year_wc'].notnull(), 'is_wc_yr'] = 'yes'
df.drop('Year_wc', axis=1, inplace=True)
df['Country_wc'] = df['Country_wc'].fillna('other')
df['Winner_wc'] = df['Winner_wc'].fillna('other')
df['Runners-Up_wc'] = df['Runners-Up_wc'].fillna('other')
df['season_wc'] = df['season_wc'].fillna('other')
df['golden_ball_wc'] = df['golden_ball_wc'].fillna('other')
df['silver_ball_wc'] = df['silver_ball_wc'].fillna('other')

# Copa America
df = pd.merge(df,
             ca_df,
             how = 'left',
             left_on = 'season_player',
             right_on = 'season_ca')

# ca_df cleaning
df['is_ca_yr'] = 'no'
df.loc[df['Year_ca'].notnull(), 'is_ca_yr'] = 'yes'
df.drop('Year_ca', axis=1, inplace=True)
df['Champions_ca'] = df['Champions_ca'].fillna('other')
df['best_player_ca'] = df['best_player_ca'].fillna('other')
df['season_ca'] = df['season_ca'].fillna('other')

# Euro Cup
df = pd.merge(df,
             ec_df,
             how = 'left',
             left_on = 'season_player',
             right_on = 'season_ec')

# ec_df cleaning
df['is_ec_yr'] = 'no'
df.loc[df['season_ec'].notnull(), 'is_ec_yr'] = 'yes'
df['winner_ec'] = df['winner_ec'].fillna('other')
df['best_player_ec'] = df['best_player_ec'].fillna('other')
df['season_ec'] = df['season_ec'].fillna('other')

# Champions League
df = pd.merge(df,
             cl_df,
             how = 'left',
             left_on = 'season_player',
             right_on = 'season_cl')

# cl_df cleaning
df.drop('Year_cl', axis=1, inplace=True)

# Ballon D'or - this is our response variable
df = pd.merge(df,
             ballon_df,
             how = 'left',
             left_on = 'season_player',
             right_on = 'season_ballon')
```

```python
cls_to_drop = ['Rk_team', 'Squad_team', 'Country_team', 'season_team', 'season_wc', 'season_ca', 'season_ec',
              'season_cl', 'season_ballon']

df.drop(cls_to_drop, axis=1, inplace=True)
```
